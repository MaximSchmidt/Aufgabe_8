### Imports

```python
from pathlib import Path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

### 8a RGB-Bild einlesen und in ein Graustufenbild umwandeln

```python
image_path = Path("images_input/1.nature_small.jpeg")

image = Image.open(image_path)
rgb_matrix = np.array(image, dtype=np.uint8)

red = rgb_matrix[:, :, 0]
green = rgb_matrix[:, :, 1]
blue = rgb_matrix[:, :, 2]

gray_matrix = 0.21 * red + 0.72 * green + 0.07 * blue
gray_matrix = gray_matrix.astype(np.uint8)

## plt.imshow() färbt 2D Matrizen standardmäßig mit Farbskala ein also:
## plt.imshow(gray_matrix, cmap="gray", vmin=0, vmax=255)
## plt.axis("off")
## plt.show()
```

### 8b Helligkeit und Kontrast vom Graustufenbild anpassen

```python
alpha = 1.3   # Kontrast
beta = 20     # Helligkeit

gray_matrix_2 = alpha * gray_matrix + beta
gray_matrix_2 = np.clip(gray_matrix_2, 0, 255)
gray_matrix_2 = gray_matrix_2.astype(np.uint8)

plt.imshow(gray_matrix_2, cmap="gray", vmin=0, vmax=255)
plt.axis("off")
plt.show()
```
