### Imports

```python
from pathlib import Path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import time
```

### 8a RGB-Bild einlesen und in ein Graustufenbild umwandeln

```python
image_path = Path("images_input/1.nature_small.jpeg")
## image_path = Path("images_input/2.nature_medium.jpeg")
## image_path = Path("images_input/3.nature_large.jpeg")
## image_path = Path("images_input/4.nature_mega.jpeg")

# cv2.imread() loads color images in BGR format by default
image = cv2.imread(str(image_path))

start = time.perf_counter()

# Convert the BGR image to a grayscale image
# OpenCV internally applies its optimized grayscale conversion
# The resulting matrix contains one grayscale value per pixel
##  0.21 * red + 0.72 * green + 0.07 * blue =
gray_matrix = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Optional visualization of the grayscale image
## plt.imshow() färbt 2D Matrizen standardmäßig mit Farbskala ein also:
## plt.imshow(gray_matrix, cmap="gray", vmin=0, vmax=255)
## plt.axis("off")
## plt.show()
```

### 8b Helligkeit und Kontrast vom Graustufenbild anpassen

```python
alpha = 1.3 # Kontrast 
beta = 20 # Helligkeit

# Convert the BGR image to grayscale using OpenCV's built-in optimized grayscale conversion
gray_matrix_2 = cv2.convertScaleAbs(gray_matrix, alpha=alpha, beta=beta)

## plt.imshow(gray_matrix_2, cmap="gray", vmin=0, vmax=255)
## plt.axis("off")
## plt.show()
```

### 8c Histogramm der Grauwerte berechnen und darstellen

```python
# Calculate the histogram of the processed grayscale image
# [0] selects the only grayscale channel
# [256] creates one histogram bin for each possible grayscale value (0-255)
# [0, 256] defines the value range that should be included
# cv2.calcHist() returns the histogram as a 2D array with shape (256, 1)
histogram = cv2.calcHist([gray_matrix_2], [0], None, [256], [0, 256])
# flatten() converts it into a one-dimensional array with 256 entries
histogram = histogram.flatten()

end = time.perf_counter()

opencv_time = end - start
print(opencv_time)

## plt.figure(figsize=(10, 5))
## plt.bar(range(256), histogram, width=1.0)
#  plt.xlabel("Grauwert")
## plt.ylabel("Anzahl Pixel")
## plt.title("Histogramm der Grauwerte")
## plt.xlim(0, 256)
## plt.grid(axis="y", alpha=0.3)
## plt.show()
```
